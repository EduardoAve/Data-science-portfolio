```python

# --- PASO 2: Función de Inversión (Asumiendo escala 1-5) ---
def invert_scale(series):
  """Invierte una escala de 1-5 (1->5, 2->4, 3->3, 4->2, 5->1)."""
  numeric_series = pd.to_numeric(series, errors='coerce')
  return 6 - numeric_series

# --- PASO 3: Calcular las nuevas variables ---

print("Calculando variables...")

# 1. Academic Resources
for col in cols_academic_resources:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df['Academic_Resources'] = df[cols_academic_resources].mean(axis=1, skipna=True)
print("Calculado: Academic_Resources")

# 2. Performance Pressure
df['Performance_Pressure'] = pd.to_numeric(df[col_performance_pressure], errors='coerce')
print("Calculado: Performance_Pressure")

# 3. Perceived Autonomy
df_autonomy = df[cols_perceived_autonomy].copy()
for col in cols_autonomy_to_invert:
    if col in df_autonomy.columns: df_autonomy[col] = invert_scale(df_autonomy[col])
    else: print(f"Advertencia: La columna a invertir '{col}' no se encontró en df_autonomy.")
for col in df_autonomy.columns:
     if col not in cols_autonomy_to_invert: df_autonomy[col] = pd.to_numeric(df_autonomy[col], errors='coerce')
df['Perceived_Autonomy'] = df_autonomy.mean(axis=1, skipna=True)
print("Calculado: Perceived_Autonomy")

# 4. Quality of Leadership
for col in cols_quality_leadership:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df['Quality_Leadership'] = df[cols_quality_leadership].mean(axis=1, skipna=True)
print("Calculado: Quality_Leadership")

# 5. Sense of Community
for col in cols_sense_community: # Usa la lista corregida
    # La verificación ahora se hace implícita al intentar convertir
    df[col] = pd.to_numeric(df[col], errors='coerce')
df['Sense_Community'] = df[cols_sense_community].mean(axis=1, skipna=True) # Usa la lista corregida
print("Calculado: Sense_Community")

# 6. Job Satisfaction
for col in cols_job_satisfaction:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df['Job_Satisfaction'] = df[cols_job_satisfaction].mean(axis=1, skipna=True)
print("Calculado: Job_Satisfaction")

# 7. Work Motivation
for col in cols_amotivation + cols_extrinsic_social + cols_extrinsic_material + cols_introjected + cols_identified + cols_intrinsic:
     df[col] = pd.to_numeric(df[col], errors='coerce')
df['Amotivation'] = df[cols_amotivation].mean(axis=1, skipna=True)
df['Extrinsic_Social'] = df[cols_extrinsic_social].mean(axis=1, skipna=True)
df['Extrinsic_Material'] = df[cols_extrinsic_material].mean(axis=1, skipna=True)
df['Introjected_Motivation'] = df[cols_introjected].mean(axis=1, skipna=True)
df['Identified_Motivation'] = df[cols_identified].mean(axis=1, skipna=True)
df['Intrinsic_Motivation'] = df[cols_intrinsic].mean(axis=1, skipna=True)
print("Calculado: Work Motivation sub-scales")

# 8. Burnout
for col in cols_burnout:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df['Burnout_Score'] = df[cols_burnout].mean(axis=1, skipna=True)
print("Calculado: Burnout_Score")

# 9. Vulnerability to Burnout Sub-scales
# Subjective Significance
for col in cols_subjective_significance: df[col] = pd.to_numeric(df[col], errors='coerce')
df['Subjective_Significance'] = df[cols_subjective_significance].mean(axis=1, skipna=True)
# Professional Ambition
for col in cols_professional_ambition: df[col] = pd.to_numeric(df[col], errors='coerce')
df['Professional_Ambition'] = df[cols_professional_ambition].mean(axis=1, skipna=True)
# Tendency Exertion
for col in cols_tendency_exertion: df[col] = pd.to_numeric(df[col], errors='coerce')
df['Tendency_Exertion'] = df[cols_tendency_exertion].mean(axis=1, skipna=True)
# Striving Perfection
for col in cols_striving_perfection: df[col] = pd.to_numeric(df[col], errors='coerce')
df['Striving_Perfection'] = df[cols_striving_perfection].mean(axis=1, skipna=True)

# Emotional Distancing (con inversión)
df_distancing = df[cols_emotional_distancing].copy()
for col in cols_distancing_to_invert:
    if col in df_distancing.columns: df_distancing[col] = invert_scale(df_distancing[col])
    else: print(f"Advertencia: La columna a invertir '{col}' no se encontró en df_distancing.")
for col in df_distancing.columns:
     if col not in cols_distancing_to_invert: df_distancing[col] = pd.to_numeric(df_distancing[col], errors='coerce')
df['Emotional_Distancing'] = df_distancing.mean(axis=1, skipna=True)

# Resignation Tendency
for col in cols_resignation_tendencies: df[col] = pd.to_numeric(df[col], errors='coerce')
df['Resignation_Tendency'] = df[cols_resignation_tendencies].mean(axis=1, skipna=True)
# Offensive Coping
for col in cols_offensive_coping: df[col] = pd.to_numeric(df[col], errors='coerce')
df['Offensive_Coping'] = df[cols_offensive_coping].mean(axis=1, skipna=True)

# Mental Stability (con inversión)
df_stability = df[cols_mental_stability].copy()
for col in cols_stability_to_invert:
     if col in df_stability.columns: df_stability[col] = invert_scale(df_stability[col])
     else: print(f"Advertencia: La columna a invertir '{col}' no se encontró en df_stability.")
for col in df_stability.columns:
     if col not in cols_stability_to_invert: df_stability[col] = pd.to_numeric(df_stability[col], errors='coerce')
df['Mental_Stability'] = df_stability.mean(axis=1, skipna=True)

# Satisfaction Work
for col in cols_satisfaction_work: # Usa la lista corregida
    df[col] = pd.to_numeric(df[col], errors='coerce')
df['Satisfaction_Work'] = df[cols_satisfaction_work].mean(axis=1, skipna=True) # Usa la lista corregida
# Satisfaction Life
for col in cols_satisfaction_life:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df['Satisfaction_Life'] = df[cols_satisfaction_life].mean(axis=1, skipna=True)

# Social Support (con inversión)
df_support = df[cols_social_support].copy()
for col in cols_support_to_invert:
     if col in df_support.columns: df_support[col] = invert_scale(df_support[col])
     else: print(f"Advertencia: La columna a invertir '{col}' no se encontró en df_support.")
for col in df_support.columns:
     if col not in cols_support_to_invert: df_support[col] = pd.to_numeric(df_support[col], errors='coerce')
df['Social_Support'] = df_support.mean(axis=1, skipna=True)

print("Calculado: Vulnerability to Burnout sub-scales")


# --- PASO 4: Identificar TODAS las columnas originales que se usaron ---
# (Usa las listas CORREGIDAS con los nombres exactos)
columns_to_drop = list(set(
    cols_academic_resources +
    [col_performance_pressure] +
    cols_perceived_autonomy +
    cols_quality_leadership +
    cols_sense_community + # Usa la lista corregida
    cols_job_satisfaction +
    cols_amotivation + cols_extrinsic_social + cols_extrinsic_material +
    cols_introjected + cols_identified + cols_intrinsic +
    cols_burnout +
    cols_subjective_significance + cols_professional_ambition + cols_tendency_exertion +
    cols_striving_perfection + cols_emotional_distancing + cols_resignation_tendencies +
    cols_offensive_coping + cols_mental_stability + cols_satisfaction_work + # Usa la lista corregida
    cols_satisfaction_life + cols_social_support
))


# --- PASO 5: Eliminar las columnas originales ---
print(f"\nIntentando eliminar {len(columns_to_drop)} columnas originales...")
existing_cols_to_drop = [col for col in columns_to_drop if col in df.columns]
missing_cols = [col for col in columns_to_drop if col not in df.columns]

if missing_cols:
    print(f"Advertencia: Las siguientes columnas listadas para eliminar no se encontraron en el DataFrame: {missing_cols}")

print(f"Se eliminarán {len(existing_cols_to_drop)} columnas.")

df.drop(columns=existing_cols_to_drop, inplace=True, errors='raise')

print("Columnas originales eliminadas.")

# --- PASO 6: Verificar el resultado (Opcional) ---
print("\nColumnas restantes:")
print(df.columns.tolist())
print(f"\nNúmero de columnas restantes: {len(df.columns)}")
```
